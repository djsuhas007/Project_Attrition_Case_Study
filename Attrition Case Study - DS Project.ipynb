```python
# We will import LabelEncoder to convert the 'categorical' values into 'continues' for further process
from sklearn.preprocessing import LabelEncoder as le
import numpy as np
```

```python
# Trnsforming the 'categorical' values into 'continues'
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    df[column]=le().fit_transform(df[column])
```

```python
# Splitting the data into X (Independant variable) & Y (Dependant variable)
# We are predicting 'Attrition', hence, it is the 'Y' column i.e. Target variable
X = df.drop(columns='Attrition',axis=1)
Y = df['Attrition']
```

```python
X.columns
```

```python
Y
```

```python
# Using train_test_split method to split the data in training & teating set
from sklearn.model_selection import train_test_split
```

```python
# 70% data in training & 30% data in testing set
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.3, random_state=1234)
```

```python
# We will use 'Ensemble' method for classification & will use RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=1234)
```

```python
# Fit the model
rfc.fit(X_train,Y_train)
```

```python
# Training data accuracy
rfc.score(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
Y_pred = rfc.predict(X_test)
```

```python
# Creating the confusion matrix
cm = confusion_matrix(Y_test,Y_pred)
```

```python
cm
```

```python
Accuracy = accuracy_score(Y_test,Y_pred)
print("Model's testing accuracy score is : {} " .format(Accuracy))
```
